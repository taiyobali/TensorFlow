1. Classification example


```python
# load_data
import tensorflow as tf 
mnist = tf.keras.datasets.mnist 
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# preprocessing
X_train, X_test = X_train/255.0, X_test/255.0 
```

```python
# build_neural_network
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout 

model = Sequential()
model.add(Input((28,28))) # 1 instance 28,28 pixels
model.add(Flatten()) #make 1D
model.add(Dense(128, activation="relu")) #neurons and break linerity
model.add(Dense(64, activation="relu")) 
model.add(Dropout(0.2)) #regularization/overfitting fix
model.add(Dense(10)) #output
```

```python
nonsense_prediction = model(X_train[:1]).numpy() #model's prediction
nonsense_prediction
```

```python
tf.nn.softmax(nonsense_prediction).numpy() #probability of prediction
```

```python

from tensorflow.keras.losses import SparseCategoricalCrossentropy
loss_fn = SparseCategoricalCrossentropy(from_logits=True) #model's losses , wrongness
# -tf.math.log(1/10)
loss_fn(y_train[:1], nonsense_prediction)
```

```python
from tensorflow.keras.optimizers import Adam 
optimizer = Adam(learning_rate=0.01) #optimizer: update models for improve

# compile model
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
# train model
model.fit(X_train, y_train, epochs=10)
```

```python
# evaluate model
model.evaluate(X_test,y_test)
```

```python
tf.argmax(tf.nn.softmax(model(X_test[:1])).numpy(), axis=1)  #return the highest probabity index
```

2. Regression exp

```python
import pandas as pd 
df = pd.read_csv("housing.csv")
df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 

X, y = df.drop(columns=["MedHouseVal"], axis=1), df["MedHouseVal"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input((8,)),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse", metrics=["mae","r2_score"])
```

```python
model.fit(X_train_scaled, y_train, epochs=10)
```

```python
model.evaluate(X_test_scaled,y_test)
```
